```python
import pandas as pd

def averages(file_path):
    dropped_lines = 0
    titles = []
    data = []

    with open(file_path, 'r') as myFile:
        file_lines = myFile.readlines()

    for line in file_lines:
        line_split = line.strip().split(',')

        if line_split[0] == "80":
            try:
                row = list(map(float, line_split[2:]))
                data.append(row)
            except ValueError:
                dropped_lines += 1

        elif line_split[0] == "02":
            titles.extend([title.strip() for title in line_split[2:]])

    df = pd.DataFrame(data, columns=titles)
    print(f"{dropped_lines} lines were dropped")
    return df

# Example usage
file_path = 'poe_long.hml'  # Replace with the actual path
result_df = averages(file_path)
print(result_df)
```
